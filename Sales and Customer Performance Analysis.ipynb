# Sales and Customer Analysis Dashboard

This project represents sales transactions for a fictional retail company operating between 2011 and 2016. The company sells a variety of bike products across multiple categories and subcategories and serves a broad customer base segmented by demographics such as age group and gender.

## Purpose of This Analysis

The business wants to better understand:

- How overall sales performance has evolved over time

- Whether revenue growth is driven by increased order volume or higher-value purchases

- How customer purchasing behavior differs across demographic segments

## Business Problem

The core business problem this analysis seeks to answer is:

**How has the business performed year-over-year, and what does customer purchasing behavior reveal about the drivers behind revenue and profit growth?**

This problem is addressed through two interactive dashboards using Tableau:

**Sales Performance Analysis Dashboard**

The sales dashboard evaluates the revenue, profit, and order quantity as well as category level performances. 

**Customer Behavior Analysis Dashboard**

The customer dashboard evaluates how purchasing behavior varies by demographic segment and how changes in order value contribute to business outcomes.

## Questions the Sales Dashboard Answers

### Question 1: How did total revenue perform compared to the previous year?

If we are comparing the current year (2016) to the previous year (2015) revenue, total revenue reached \\$17M, representing a 11.54\% decrease year-over-year compared to 2015 where the total revenue was \\$20M. 

Revenue growth diminished in the second half of the year starting in July after the mid-year where there were no sales from July until the reast of the year. 

Despite incomplete data for 2016, the upward trend in 2015 suggests a strong sales momentum. 

### Why are there no records for Q3 and Q4 in 2014 and 2016?

The dataset does not contain records for the final five months in 2014 and 2016. Because of this, year-over-year analysis involiving these years should be taken with caution particularly when looking towards the end of the year.

To provide additional context, external economic conditions during these periods were researched to better understand the broader business environment in which the data exists. This contextual research is used to support interpretation, not to directly explain or seen as an attribute to the missing data.

### 2014 
It looks like businesses experienced slower growth, particularily in Europe which was still trying to recover from prior financial challenges. People were not spending as much especially on non-essential items. And while growth across businesses weresteady, it was not strong.

### 2016 
In 2016, economic and political uncertainty increased due to events such as the Brexit referendum in Europe and the U.S. presidential election. During periods of uncertainty, consumers often delay larger or discretionary purchases, which can lead to uneven performance across the year for many businesses.


### Question 2: Is revenue growth translating to profit growth?

As we continue to look at the revenue and compare how much money is being profitted it can be said that there is a correlation between the two. Over the course of 2011 to 2016, when revenue increased or decreased, profit followed in the same trajectory of decreasing or increasing as well. So that means that profits were stable during the periods where revenue increased. 

An example of this is that in 2015 total revenue reached \\$20M which was a 41.49\% increase from the year before. Profit followed the same increase with a total of \\$7.5M in profit which was a 28.38\% increase from the previous year.  

### Question 3: Which product contributed most to revenue and profit?

Road Bikes and Mountain Bikes are the two top products that remain consistent in contributing to the revenue and profit which makes them important to the business. 

### Question 4: How do revenue and profit perform relative to average benchmarks over time?

Look at each year closely, starting from 2013 when the business started to pick up momentum, it can be stated that revenue and profit exceed the average benchmarks during the latter part of the year. 

## Questions the Customers Dashboard Answers

This section will be a summary of my findings for each question the business wanted answers to and what these findings mean for their customers.

**Important to Note**: The dataset that I am using does not include unique customer identifiers such as a "Customer ID". Because of this, customer-related metrics are analyzed at the order level rather than at the individual customer level. Metrics such as total customers, customer lifetime value, or retention are intentionally excluded to maintain analytical accuracy for what the dataset provides. 

### Question 1: How active are customers in terms of order volume?

Total orders fluctuated over the years and so did the revenue and average order value. However, there were some years where the total orders decreased but the revenue increased. For example, in 2015 total orders dropped by 16\% over the previous year. But the total revenue increased which indicates that customers were shifting towards higher value products. 

### Question 2: How much are customers spending per order on average?

In 2015, the average order value increased by 85\% with an average order value of `$69.20` versus the previous year. However, the total orders decreased by 16\% but the total revenue still took an upward trajectory increasing by 41\%. This tells us that the revenue growth is being driven more so by how much customers are buying rather than how many times they are ordering products.  

### Question 3: Which age groups contributed the most to revenue growth?

If we are looking at years 2015 and 2016 on a year to year basis, adults (35-64) and young adults (25-34) are those who are driving revenue for the business in both years. In 2016 adults who purchased products increased by `$1M`. 

Young adults decreased by `$2M` but were still the second largest age group that drove revenue that year.

### Question 4: How does revenue differ by gender?

There is a balance between male and female customers as they drive revenue. In 2015 there wasn't much of a difference in how much revenue both groups generated for the business. The revenue was split in half in 2015 where both genders brought in \$10M. 

In 2016 there was a bit of a decrease where females generated `$8.3M`. Males brought in `$9.5M` which they are still scaled closely together.
